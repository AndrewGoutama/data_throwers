## Removing unwanted Columsn and NA Datas

```r
## Removing start_lat, start_lng, end_lat, end_lng Columns
## Removing NA Data with na.omit()
data2024 <- data2024 %>%
  select(-c(start_lat, start_lng, end_lat, end_lng)) %>%
  na.omit()

# checking the dataset for any missing values
colSums(is.na(data2024))

##Check and remove any duplicate values
paste("Duplicate Values :", sum(duplicated(data2024$ride_id)))

##Check if there is any TEST or DEMO station names
data2024[grep("TEST", data2024$start_station_name, ignore.case = TRUE),]
data2024[grep("TEST", data2024$end_station_name, ignore.case = TRUE),]
data2024[grep("DEMO", data2024$start_station_name, ignore.case = TRUE),]
data2024[grep("DEMO", data2024$end_station_name, ignore.case = TRUE),]
```

## Adding New Columns and Datas
Additional variables willl be created for better analysis of the dataset. The new variables include:

##### all_stations_used: combining the start and end stations of each ride
##### ride_length: Calculate the length of the ride and convert it into minutes
##### day_of_week: Includes the day of the week for each rides
##### month: month of the year for each rides


```r
#Station Used [all_stations_used]
data2024 <- unite(data2024, all_stations_used, start_station_name, 
                  end_station_name, sep = " - ")
```

```r
#Total time taken [ride_length]
data2024$ride_length <- difftime(data2024$ended_at, data2024$started_at)

#Make sure the [ride_length] isn't negative
print(head(data2024$ride_length))
data2024 <- filter(data2024, ride_length>0) 
```

```r
## Making the format into Date Time
data2024$started_at <- as.POSIXct(data2024$started_at, 
                                  format = "%Y-%m-%d %H:%M:%S")
data2024$ended_at <- as.POSIXct(data2024$ended_at, 
                                format = "%Y-%m-%d %H:%M:%S") 

## adding the [day of the week] besides started_at
data2024 <- add_column(data2024, day_of_week = weekdays(data2024$started_at), 
                       .after = "started_at")
```

```r
# Make sure again that 'started_at' is in Date format
data2024$started_at <- as.Date(data2024$started_at)

# Adding the 'month' column with full month names
data2024 <- add_column(data2024, month = format(data2024$started_at, "%B"), 
                       .after = "day_of_week")

# Converting 'day_of_week' and 'month' data into more casual names
data2024$day_of_week <- ordered(data2024$day_of_week, 
  levels = c("Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"))
data2024$month <- ordered(data2024$month, 
  levels = c("January", "February", "March", "April", "May", "June",
             "July", "August", "September", "October", "November", "December"))
```

# STAGE 4 ANALYZE
##### Analyzing the data in the dataset

Using statistical methods to find patterns, relationships and trends in the data. The primary objective of this phase is to figure out the behavioural pattern of cyclist members and non-members with regards to their riding habits.

```r
#Summary to see the summary of the dataset
summary(data2024$ride_length)
```

### Calculations

```r
# Creating a Custom Function called 'mode' that will find out most frequent value
mode <- function(x){
  unique_x <- unique(x)
  unique_x[which.max(tabulate(match(x, unique_x)))]
}
```

```r
# Finding out the member-wise average, maximum and mode of ride_length
## [Average of Member/Casual], [Maximum], Mode[Ride Length]

ride_summary <- data2024 %>%
  group_by(member_casual) %>%
  summarise(Total_Rides = n(),
            avg_ride_length = mean(ride_length), 
            max_ride_length = max(ride_length), 
            mode_ride_length = mode(ride_length)) %>%
  rename(membership = member_casual)

## Print ride_summary to see the results
print(ride_summary) 
```

```r
# calculating the member-wise, [ride-type total] count as well as the [average ride length]
# [ride-type total], [average ride length]
ride_type_count <- data2024 %>%
  group_by(member_casual, rideable_type) %>%
  summarise(ride_count = n(), avg_ride_length = mean(ride_length), .groups = "drop") %>%
  rename(membership = member_casual)

print(ride_type_count) 
```

```r
# calculating the member-wise, day-wise ride count and average ride length and based on weekend or weekday
rides_by_day <- data2024 %>%
  group_by(member_casual, day_of_week) %>% 
  summarise(ride_count = n(), avg_ride_length = mean(ride_length), .groups = "drop") %>%
  rename(membership = member_casual) %>%
  arrange(membership, desc(ride_count)) %>%
  mutate(day_category = ifelse(day_of_week %in% c("Saturday", "Sunday"), "Weekend", "Weekday"))

# printing the results in descending order based on ride_count
print(rides_by_day) 
```

```r
# calculating the month-wise, member-wise ride count and average ride length
rides_per_month <- data2024 %>%
  filter(!is.na(month)) %>%
  group_by(member_casual, month) %>%
  summarise(ride_count = n(), avg_ride_length = mean(ride_length), .groups = "drop") %>%
  rename(membership = member_casual) %>%
  arrange(membership, desc(ride_count))

print(rides_per_month, n = nrow(rides_per_month))
```

```r
# finding out the popular stations used by casual members based on where riders 
#start and end their rides more frequently
popular_stations <- data2024 %>%
  filter(member_casual == "casual") %>% 
## to find ones for normal members just switch 'casual' with 'member'
  group_by(all_stations_used) %>%
  summarise(ride_count = n(), .groups = "drop") %>%
  arrange(desc(ride_count)) %>%
  head(10)

#Most Popular Stations on 2024
cat("Top 10 most popular stations used by casual members :\n\n")
print(popular_stations)
```

# STAGE VI SHARE
##### Sharing insights on the data with stakeholders



Recommendations for Cyclistic's Marketing Strategy

Based on the insights derived from the analysis, here are tailored recommendations to convert casual riders into annual members and improve engagement:

1. Create Targeted Membership Packages

- Offer Memberships: Introduce weekend-based memberships that appeal to casual riders who prefer riding during specific days.

- Highlight Cost Savings and Convenience: Since casual riders tend to have longer ride times (average of 25 minutes), emphasize the potential cost savings they can achieve with an annual membership versus pay-as-you-go.

- Tiered Membership Options: Create tiered plans that provide benefits based on usage, such as priority reservations for classic bikes, which are popular among both casuals and members.

2. Promote Weekend Events
- Organize Weekend-Exclusive Events: As casual riders tend to ride more on weekends, Cyclistic could offer weekend-only events or promotions. For example, offer double rewards points for rides taken on weekends or exclusive access to popular landmarks like Navy Pier. 
- Collaborate with Local Attractions: Partner with nearby attractions around top stations like Streeter Dr & Grand Ave and DuSable Lake Shore Dr & Monroe St to offer discounts to annual members, creating a combined leisure experience that encourages membership upgrades.

3. Capitalize on Ride Duration Patterns
- Seasonal Promotions for Summer Break: Given the high ridership in July, August, and September, Cyclistic should run summer campaigns encouraging long-term memberships with added perks during these months, such as free family rides or discounted group memberships.

- Time-Limited Offers Based on Ride Frequency: Since casual riders have longer average ride times, offer incentives for reaching certain ride milestones (e.g., free ride after 5 rides of 25+ minutes) to further engage them and highlight the benefits of membership.

4. Increase Awareness Through Targeted Communication
- Personalized Messaging: Use email and app notifications to share personalized ride statistics with casual riders, like their ride duration and money saved through membership options.

- Promote Classic and Electric Bikes Benefits: Since classic and electric bikes are popular, focus on promoting exclusive features like availability guarantees, priority bookings, and special discounts on these bike types for members.

5. Enhance Visibility Through Station-Based Campaigns
- Promote at Popular Tourist Stations: Since Streeter Dr & Grand Ave and DuSable Lake Shore Dr & Monroe St are popular locations with tourists, create station-based ads that encourage casual riders to upgrade for benefits like extended ride durations.

- Partner with Local Tourism Organizations: Collaborate with tourism entities around these stations to offer joint promotions for tourists, making an annual membership more appealing for frequent visitors or those on extended vacations.

These recommendations leverage current trends and user behavior, aiming to create a targeted approach that not only converts casual riders into annual members but also enhances the overall Cyclistic experience. This approach will help drive user retention, growth, and engagement, supporting Cyclistic's long-term objectives.
